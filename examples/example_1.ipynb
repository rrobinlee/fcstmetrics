## Example 1 

```python
import pandas as pd
import numpy as np
from fcstmetrics import (
    validate_forecast,
    convert_to_sktime_format,
    prepare_sktime_train_test
)
import warnings
warnings.filterwarnings("ignore")
```

```python
np.random.seed(42)
n_train, n_test = 100, 20
t = np.arange(n_train + n_test)
trend = 100 + 2 * t
seasonality = 20 * np.sin(2 * np.pi * t / 12)
noise = np.random.randn(n_train + n_test) * 5
y_actual = trend + seasonality + noise
```

```python
y_pred = trend + seasonality + np.random.randn(n_train + n_test) * 4

y_train = y_actual[:n_train]
y_test = y_actual[n_train:]
y_train_pred = y_pred[:n_train]
y_test_pred = y_pred[n_train:]
```

```python
report = validate_forecast(y_train=y_train,
                           y_train_pred=y_train_pred,
                           y_test=y_test,
                           y_test_pred=y_test_pred,
                           model_name="Example  Model",
                           print_report=True,
                           check_stationarity=True)
```

```python
print(f"\nTest Set Performance:")
print(f"  RMSE: {report.metrics['test_RMSE']:.4f}")
print(f"  MAPE: {report.metrics['test_MAPE']:.2f}%")
print(f"  R2:   {report.metrics['test_R2']:.4f}")
print(f"  Directional Accuracy: {report.metrics['test_Directional_Accuracy']:.2f}%")
```
